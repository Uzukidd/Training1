#### Exercise : size analysis of iris flower

Requirement：

	1. Size relationship between sepal and petal (scatter diagram)

   	2. The size relationship between sepals and petals of iris of different species
   	3. Distribution of sepals and petal sizes of different Iris species (box diagram)

dataSet：

```python
import seaborn as sns

data = sns.load_dataset("iris")
```

Method to spawn linearRegression

```python
def linearRegress(x, y) :

    X_sum = x.sum();

    X2_sum = (x * x).sum()

    Y_sum = y.sum()

    XY = (x * y).sum()

    n = x.shape[0]
    
    k = (X_sum * Y_sum/n - XY) / (X_sum ** 2 / n - X2_sum)

    b = (Y_sum - k * X_sum) / n

    return (k, b)
```

```python
print(data['species'].value_counts())
```

```python
data['sepal_size'] = data['sepal_length'] * data['sepal_width']

data['petal_size'] = data['petal_length'] * data['petal_width']

data = data.sort_values(['sepal_size', 'petal_size'])

############# Sort in sepal size order ###############


setosa = data[data['species'] == 'setosa']

virginica = data[data['species'] == 'virginica']

versicolor = data[data['species'] == 'versicolor']
```

```python
x = virginica['sepal_size']

y = virginica['petal_size']

k, b = linearRegress(x, y)

li_x = np.arange(x.min(), x.max() + 1)

li_y = li_x * k + b

plt.ylabel('petal size')

plt.xlabel('sepal size')

plt.title('virginica')

plt.scatter(x, y, c = '#FF0000')

plt.plot(li_x, li_y, c = '#000000')

plt.show()
```

```python
x = versicolor['sepal_size']

y = versicolor['petal_size']

k, b = linearRegress(x, y)

li_x = np.arange(x.min(), x.max() + 1)

li_y = li_x * k + b

plt.ylabel('petal size')

plt.xlabel('sepal size')

plt.title('versicolor')

plt.scatter(x, y, c = '#FF0000')

plt.plot(li_x, li_y, c = '#000000')

plt.show()
```

```python
x = setosa['sepal_size']

y = setosa['petal_size']

k, b = linearRegress(x, y)

li_x = np.arange(x.min(), x.max() + 1)

li_y = li_x * k + b

plt.ylabel('petal size')

plt.xlabel('sepal size')

plt.title('setosa')

plt.scatter(x, y, c = '#FF0000')

plt.plot(li_x, li_y, c = '#000000')

plt.show()
```

x : sepal_size     y : petal_size

```python
relationship = lambda x, K, B : x * K + B
```

```python
k = virginica.shape[0]

x = virginica['sepal_size']

y = virginica['petal_size']

K, B = linearRegress(x, y)

y = relationship(virginica['sepal_size'], K, B)

plt.plot(x, y, color = '#8000FF', label = 'virginica')

############# Virginica ###############



k = versicolor.shape[0]

x = versicolor['sepal_size']

y = versicolor['petal_size']

K, B = linearRegress(x, y)

y = relationship(versicolor['sepal_size'], K, B)

plt.plot(x, y, color = '#AA2282', label = 'versicolor')

############# Versicolor #############




k = setosa.shape[0]

x = setosa['sepal_size']

y = setosa['petal_size']

K, B = linearRegress(x, y)

y = relationship(setosa['sepal_size'], K, B)

plt.plot(x, y, color = '#800030', label = 'setosa')

############# Setosa #################



plt.ylabel('petal size')

plt.xlabel('sepal size')

plt.legend()

plt.show()
```

```python
datas = [virginica['sepal_size'], versicolor['sepal_size'], setosa['sepal_size']]
 
plt.boxplot(datas, sym = 'rs')

plt.xticks([1, 2, 3], ['virginica', 'versicolor', 'setosa'])

plt.ylabel('sepal size')

plt.title('Box diagram of sepal size for 3 species')

plt.show()
```

```python
datas = [virginica['petal_size'], versicolor['petal_size'], setosa['petal_size']]
 
plt.boxplot(datas, sym = 'rs')

plt.xticks([i + 1 for i in range(3)], ['virginica', 'versicolor', 'setosa'])

plt.ylabel('petal size')

plt.title('Box diagram of petal size for 3 species')

plt.show()
```
