```python
targetDB = "iris"
```

Connect the DBS mysql.

```python
sql = mysql.connect(
    host = "localhost",
    password = "", #Please input your password here
    user = "root",
    charset = "utf8")

cursor = sql.cursor()
```

```python
def executeSql(cmd, sql) :
    try:
        cursor = sql.cursor()
        
        cursor.execute(cmd)
        
        results = cursor.fetchall()
        
        return results
    
    except Exception as e:
        
        print(e)
        
        print("When executing")
        
        sql.rollback()
        
def executeManySql(cmd, sql, data) :
    try:
        cursor = sql.cursor()
        
        cursor.executemany(cmd, data)
        
        sql.commit()
        
        results = cursor.fetchall()
        
        return results
    
    except Exception as e:
        
        print(e)
        
        sql.rollback()
        
```

```python
def init(data, sql) :

    executeSql("create database " + targetDB + ";", sql)
    
    executeSql("use " + targetDB + ";", sql)

    executeSql("""
    CREATE TABLE flower (
    id INT auto_increment PRIMARY KEY , 
    sepal_length FLOAT NOT NULL, 
    sepal_width FLOAT NOT NULL, 
    petal_length FLOAT NOT NULL, 
    petal_width FLOAT NOT NULL, 
    species CHAR(16) NOT NULL
    )
    """, sql)
    
    executeManySql("""
    INSERT INTO flower(sepal_length, sepal_width, petal_length, petal_width, species)
    VALUES(%s, %s, %s, %s, %s)
""", sql, [tuple(xi) for xi in data.values])
```

```python
count = len(executeSql("show databases like '" + targetDB + "';", sql))

if count == 0 :
    
    init(data, sql)

executeSql("use " + targetDB + ";", sql)
```

SELECT语句，无差别输出所有数据。

```python

results = executeSql("SELECT * FROM flower;", sql)

for row in results:

  print ("id:%4d, sepal_length:%f, sepal_width:%f, petal_length:%f, petal_width:%f, species:%s" % \
          (row[0], row[1], row[2], row[3], row[4], row[5]))
```

UPDATE语句，对花种进行汉化。

```python
executeSql("UPDATE flower SET species = '山鸢尾' WHERE species = 'setosa';", sql)

executeSql("UPDATE flower SET species = '杂色鸢尾' WHERE species = 'versicolor';", sql)

executeSql("UPDATE flower SET species = '维吉尼亚鸢尾' WHERE species = 'virginica';", sql)


results = executeSql("SELECT * FROM flower;", sql)
for row in results:

  print ("id:%4d, sepal_length:%f, sepal_width:%f, petal_length:%f, petal_width:%f, species:%s" % \
          (row[0], row[1], row[2], row[3], row[4], row[5]))
```

ORDER BY语句，以sepal_length进行排序

```python
results = executeSql("SELECT * FROM flower ORDER BY sepal_length;", sql)
for row in results:

  print ("id:%4d, sepal_length:%f, sepal_width:%f, petal_length:%f, petal_width:%f, species:%s" % \
          (row[0], row[1], row[2], row[3], row[4], row[5]))
```

DELETE语句，删除所有"杂色鸢尾"。

```python
executeSql("DELETE FROM flower WHERE species = '杂色鸢尾'", sql)


results = executeSql("SELECT * FROM flower;", sql)
for row in results:

  print ("id:%4d, sepal_length:%f, sepal_width:%f, petal_length:%f, petal_width:%f, species:%s" % \
          (row[0], row[1], row[2], row[3], row[4], row[5]))
```
